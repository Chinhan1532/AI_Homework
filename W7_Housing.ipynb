```python
import pandas as pd
url = '/content/Housing.csv'
data = pd.read_csv(url)
data.head()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

data.hist()
plt.show()
     
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter',x='price',y='area',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model= linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['price']), y=data['area'])
print('Gi tr Intercept l:',model.intercept_)
print('Gi tr Coeffient l:',model.coef_)

model.score(X=pd.DataFrame(data['price']), y=data['area'])
```

```python
data.plot(kind='scatter',x='price',y='area',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['price']))
plt.plot(data['price'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['price'], data['price']**2]).T #Tranpose: chuyn v
model.fit(X, y = data['area'])
print('H s Intercept:', model.intercept_)
print('H s Coeffient:', model.coef_)
print(' chnh xc ca m hnh:', model.score(X, y = data['area']))
```

```python
import numpy as np
range = np.arange(0.1, 1.5, 0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'price', y = 'area', figsize = (9,9), color = 'blue', xlim = (1,1.5))
plt.plot(range, y_pred, color = 'red')
```
